```python
# Dependencies and Setup
import pandas as pd
import os

# File to Load 
school_data_to_load = os.path.join('Resources', 'schools_complete.csv')
student_data_to_load = os.path.join('Resources', 'students_complete.csv')


# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Create a new dataframe and calculate the summary stats.
district_summary_df = pd.DataFrame({'Total Schools' : len(school_data_complete['school_name'].unique()), 
                                    'Total Students':school_data_complete['school_name'].count(),
                                    'Total Budget' :school_data['budget'].sum(),
                                    'Average Math Score' :school_data_complete['math_score'].mean(),
                                    'Average Reading Score':school_data_complete['reading_score'].mean(),
                                    '% Passing Math':(school_data_complete["math_score"] >= 70).sum()/school_data_complete["math_score"].count(),
                                    '% Passing Reading': (school_data_complete["reading_score"] >= 70).sum()/school_data_complete["reading_score"].count(),
                                    '% Overall Passing':(((school_data_complete['math_score'] >= 70) & (school_data_complete['reading_score'] >= 70)).sum())/school_data_complete['school_name'].count()}, 
                                   index =[0])
#format and display
pd.options.display.float_format = '{:,.2%}'.format
district_summary_df['Total Students'] = district_summary_df['Total Students'].map('{:,.0f}'.format)
district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map('${:,.0f}'.format)
district_summary_df['Average Math Score'] = district_summary_df['Average Math Score'].map('{:,.2f}'.format)
district_summary_df['Average Reading Score'] = district_summary_df['Average Reading Score'].map('{:,.2f}'.format)
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#different method. calculate everything discreetly first.
grouped_schools_df = school_data_complete.groupby(['school_name'])
types_of_school_list = school_data.set_index('school_name')['type']
students_per_school = grouped_schools_df['Student ID'].count()
budget_per_school = school_data.set_index('school_name')['budget']
budget_per_student = school_data.set_index('school_name')['budget']/school_data.set_index('school_name')['size']
average_math = grouped_schools_df['math_score'].mean()
average_reading = grouped_schools_df['reading_score'].mean()
pct_pass_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['Student ID'].count()/students_per_school 
pct_pass_read = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['Student ID'].count()/students_per_school 
pct_pass_both = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/students_per_school 
#set up school summary
school_summary_df = grouped_schools_df[['school_name']].count()
school_summary_df ['School Type'] =  types_of_school_list
school_summary_df ['Total Students'] = students_per_school
school_summary_df ['Total School Budget'] = budget_per_school
school_summary_df ['Per Student Budget'] = budget_per_student
school_summary_df ['Average Math Score'] = average_math
school_summary_df ['Average Reading Score'] = average_reading
school_summary_df ['% Passing Math'] = pct_pass_math
school_summary_df ['% Passing Reading'] = pct_pass_read
school_summary_df ['% Overall Passing'] = pct_pass_both
#format, tidy and display
del school_summary_df['school_name']
pd.options.display.float_format = '{:,.2%}'.format
school_summary_df['Total Students'] = school_summary_df['Total Students'].map('{:,.0f}'.format)
school_summary_df['Total School Budget'] = school_summary_df['Total School Budget'].map('${:,.0f}'.format)
school_summary_df['Per Student Budget'] = school_summary_df['Per Student Budget'].map('${:,.0f}'.format)
school_summary_df['Average Math Score'] = school_summary_df['Average Math Score'].map('{:,.2f}'.format)
school_summary_df['Average Reading Score'] = school_summary_df['Average Reading Score'].map('{:,.2f}'.format)
school_summary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#create a new series for each grade. use .loc on the original merged data set. Filter by grade and then groupby school.
grade_nine = school_data_complete.loc[school_data_complete['grade'] == '9th'].groupby('school_name')["math_score"].mean()
grade_ten = school_data_complete.loc[school_data_complete['grade'] == '10th'].groupby('school_name')["math_score"].mean()
grade_eleven = school_data_complete.loc[school_data_complete['grade'] == '11th'].groupby('school_name')["math_score"].mean()
grade_twelve = school_data_complete.loc[school_data_complete['grade'] == '12th'].groupby('school_name')["math_score"].mean()

math_scores_by_grade = pd.DataFrame({
        "9th": grade_nine,
        "10th": grade_ten,
        "11th": grade_eleven,
        "12th": grade_twelve})
pd.options.display.float_format = '{:,.2f}'.format
math_scores_by_grade
```

## Reading Score by Grade

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Create bins and group names. Cut the school data dataframe as a new column, then group that column in a new dataframe.
bins = [0, 584.9, 629.9, 644.9, 680]
group_name = ['< $584', "$585 - 629", "$630 - 644", "$645-675"]
school_data_complete['Spending Bucket'] = pd.cut(school_data_complete['budget']/school_data_complete['size'], bins, labels = group_name)

#group the original dataset by Spending Bucket
grouped_spend_df =  school_data_complete.groupby(['Spending Bucket'])

#calculate
math_average = grouped_spend_df['math_score'].mean()
reading_average = grouped_spend_df['reading_score'].mean()
math_pass = school_data_complete[school_data_complete['math_score'] >= 70].groupby('Spending Bucket')['Student ID'].count()/grouped_spend_df['Student ID'].count()
read_pass = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('Spending Bucket')['Student ID'].count()/grouped_spend_df['Student ID'].count()
math_read_pass = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('Spending Bucket')['Student ID'].count()/grouped_spend_df['Student ID'].count()

#format and display
scores_by_school_spending = pd.DataFrame({
    "Average Math Score": math_average,
    "Average Reading Score": reading_average,
    '% Passing Math': math_pass,
    '% Passing Reading': read_pass,
    "Overall Passing Rate": math_read_pass
            
})
scores_by_school_spending
    
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#Create bins and group names. Cut the school data dataframe as a new column, then group that column in a new dataframe.
bins = [0, 999, 1999, 4999]
group_name = ['Small (<1000)', "Medium (1000-2000)", "Large (2000-5000)"]
school_data_complete['Size Bucket'] = pd.cut(school_data_complete['size'], bins, labels = group_name)

#group the original dataset by Size Bucket
grouped_size_df =  school_data_complete.groupby(['Size Bucket'])

#calculate
math_average = grouped_size_df['math_score'].mean()
reading_average = grouped_size_df['reading_score'].mean()
math_pass = school_data_complete[school_data_complete['math_score'] >= 70].groupby('Size Bucket')['Student ID'].count()/grouped_size_df['Student ID'].count()
read_pass = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('Size Bucket')['Student ID'].count()/grouped_size_df['Student ID'].count()
math_read_pass = school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('Size Bucket')['Student ID'].count()/grouped_size_df['Student ID'].count()

#format and display
scores_by_school_size = pd.DataFrame({
    "Average Math Score": math_average,
    "Average Reading Score": reading_average,
    '% Passing Math': math_pass,
    '% Passing Reading': read_pass,
    "Overall Passing Rate": math_read_pass})
scores_by_school_size
```
